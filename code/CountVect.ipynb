**Created a function to look at the x most common words of each subreddit**

```python
#imports
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer

#to return bigrams, set num = 2, to return trigrams, set num = 3
#set type equal to either dataframe or graph
def get_common_words(path, num, word_count, type):
    df = pd.read_csv(path).drop(columns='Unnamed: 0')
    cv = CountVectorizer(stop_words='english', ngram_range=(num,num))
    data = pd.DataFrame(cv.fit_transform(df['title']).A, columns=cv.get_feature_names_out())
    
    if type=='dataframe':
        return pd.DataFrame(data.sum().sort_values(ascending=False).head(word_count))
    elif type=='graph':
        return data.sum().sort_values(ascending=False).head(15).plot(kind='barh');
```

```python
#baseball

baseball = get_common_words('./outputs/baseball-clean.csv', 2, 15, 'graph')
baseball;
```
